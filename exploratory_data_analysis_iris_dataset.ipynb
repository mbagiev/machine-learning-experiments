# Exploratory Data Analysis of Iris Dataset

```python
%matplotlib inline

import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,5))
fig.tight_layout()
fig.subplots_adjust(wspace = 0.25)
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm", ax=ax1)
iris.plot(kind="scatter", x="PetalLengthCm", y="PetalWidthCm", ax=ax2)
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=8)
```

```python
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris, size=8)
```

```python
sns.FacetGrid(iris, hue="Species", size=8).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

```python
sns.FacetGrid(iris, hue="Species", size=8).map(plt.scatter, "PetalLengthCm", "PetalWidthCm").add_legend()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))
fig.tight_layout()
fig.subplots_adjust(wspace = 0.25)
sns.boxplot(x="Species", y="SepalLengthCm", data=iris, ax=ax1)
sns.boxplot(x="Species", y="SepalWidthCm", data=iris, ax=ax2)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
fig.tight_layout()
fig.subplots_adjust(wspace = 0.25)
sns.boxplot(x="Species", y="PetalLengthCm", data=iris, ax=ax[0])
sns.boxplot(x="Species", y="PetalWidthCm", data=iris, ax=ax[1])
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))
fig.tight_layout()
fig.subplots_adjust(wspace = 0.25)
sns.violinplot(x="Species", y="SepalLengthCm", data=iris, ax=ax1)
sns.violinplot(x="Species", y="SepalWidthCm", data=iris, ax=ax2)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
fig.tight_layout()
fig.subplots_adjust(wspace = 0.25)
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, ax=ax[0])
sns.violinplot(x="Species", y="PetalWidthCm", data=iris, ax=ax[1])
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", diag_kind="kde", size=3)
```

## Correlation matrix heatmap

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(12, 10))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Draw the heatmap
sns.heatmap(iris.drop("Id", axis=1).corr(), vmax=1, square=True, cmap=cmap, linewidths=0, cbar_kws={"shrink": 0.5}, ax=ax)
```
